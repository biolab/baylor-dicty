All embeddings are based on non-averaged RPKUM data and for visualisation the averaged RPKUM data is transformed with the previously created embedding. Embeddings are either based on data of all strains ('all') or AX4 only ('AX4'). Genes (features) were preselected with ImpulseDE2 to find genes that smoothly vary acorsss developmental stages in AX4 and are thus potential developmental genes. The used features are either all potentially developmental genes identified with ImpulseDE2 ('variable') or top 2000 genes according to ImpulseDE2 (impulse fit) ('top').

```python
log=True
scale='m0s1'
ref_strains='AX4'
ref_genes='variable'

data_reference,scaler=ref_data[(ref_strains,ref_genes)]
print(ref_strains,ref_genes,data_reference.shape)

#Scale reference
data_reference = pd.DataFrame(scaler.transform(data_reference,log=log,scale=scale), 
                              index=data_reference.index, columns=data_reference.columns)
# PCA
pca = PCA(n_components=1, random_state=0)
pca.fit(data_reference)
```

```python
#Splines
for strain in conditions['Strain'].unique():
    


    conditions_strain=conditions.query('Strain =="'+strain+'"')
    times=conditions_strain['Time']
    replicates=conditions_strain['Replicate']
    data_strain = genes[conditions_strain['Measurment']].T[data_reference.columns]
    print(strain,data_strain.shape)
    data_strain = pd.DataFrame(scaler.transform(data_strain,log=log,scale=scale), index=data_strain.index, 
                                       columns=data_strain.columns)
    pca_transformed = pca.transform(data_strain).ravel()
    data_transformed=pd.DataFrame({'x':times,'y':pca_transformed})
    data_transformed=data_transformed.sort_values('x')
    var_sum=0
    for x,ys in data_transformed.groupby('x'):
        var_sum+=ys['y'].var()*ys.shape[0]

    # Base knots on expected sum of variances at each x multiplied with scaling factor s
    # It was empirically determined that s are usually below 1 for the dataset
    # Start with s > 0 and slowly proceed upwards to find where predictions start 
    # (if s is too low nothing is predicted) 
    # and where n knots becomes 2 (e.g. cubic fit); (termed first_present and first_two, respectively)
    # Find 10 locations between these two values for cross validation
    s=0.1
    x_query=data_transformed['x'].min()
    first_present=None
    first_two=None
    s_list=None

    while first_two is None:
        spline=UnivariateSpline(data_transformed['x'],data_transformed['y'],k=3,
                               s=var_sum*s)
        #Preediction not nan
        if not np.isnan(float(spline(x_query))):
            knots=len(spline.get_knots())
            if knots>2 and first_present is None:
                first_present=s
            elif knots==2:
                first_two=s
                break
        s=round(s+0.01,2)
    #s=first_two
    #not_na=True
    #while not_na:
    #    spline=UnivariateSpline(data_transformed['x'],data_transformed['y'],k=3,
    #                          s=var_sum*s)
        #Preediction not nan
    #    if not np.isnan(float(spline(x_query))):
    #        first_present=s
    #        s=round(s-0.01,2)
    #    else:
    #        not_na=False     
    if first_present is None:
        raise Error('First present is None')
    s_list=np.linspace(first_present,first_two,10)
    #s_list=list(s_list)+[first_two+1]

    print(s_list)

    #Cross validate - loo
    splitter=LeaveOneOut()
    squarred_errors=defaultdict(list)
    for train_index, test_index in splitter.split(data_transformed):
        data_train, data_test = data_transformed.iloc[train_index,:], data_transformed.iloc[test_index,:]
        var_sum=0
        for x,ys in data_train.groupby('x'):
            samples=ys.shape[0]
            if samples>1:
                var_sum+=ys['y'].var()*samples

        for s in s_list:
            spline=UnivariateSpline(data_train['x'],data_train['y'],k=3,
                                       s=var_sum*s)
            prediction=float(spline(data_test['x']))
            if not np.isnan(prediction):
                squared_error=(float(data_test['y'])-prediction)**2
                squarred_errors[s].append(squared_error)
            else:
                squarred_errors[s].append(prediction)

    # select best s
    mese=pd.DataFrame()
    for s in s_list:
        sqes=squarred_errors[s]
        n=len(sqes)
        # To account for extreme points in some samples use median instead of mean
        me=np.nanmedian(sqes)
        nan=np.isnan(sqes).sum()/n
        seme=np.nanstd(sqes)/(n**0.5)
        print(s,len(sqes),me,seme,me+seme,nan)
        mese=mese.append({'s':s,'mese':me,'nan':nan},ignore_index=True )

    has_samples=mese.query('nan < 0.3')
    if has_samples.shape[0]<1:
        has_samples=mese
        
    best=has_samples.sort_values('mese').iloc[0,:]
    print(best)
    s=float(best['s'])
    #print(mese.sort_values('mese').iloc[0,:])

    var_sum=0
    for x,ys in data_transformed.groupby('x'):
        samples=ys.shape[0]
        if samples>1:
            var_sum+=ys['y'].var()*samples


    spline=UnivariateSpline(data_transformed['x'],data_transformed['y'],k=3,
                               s=var_sum*s)
    xs = np.linspace(min(times),max(times), 1000)
    fig,ax=plt.subplots()
    ax.plot(xs, spline(xs), 'g', lw=3)
    replicates_unique=list(replicates.unique())
    cmap=plt.get_cmap('tab10').colors[:len(replicates_unique)]
    rep_colours=dict(zip(replicates_unique,cmap))
    ax.scatter(times,pca_transformed,c=[rep_colours[rep] for rep in replicates])
    ax.set_title(strain+' '+str(s)+': knots: '+str(spline.get_knots()))
    display(ax)
```

```python
#Splines
for strain in conditions['Strain'].unique():

#strain='tgrB1C1'

    conditions_strain=conditions.query('Strain =="'+strain+'"')
    times=conditions_strain['Time']
    replicates=conditions_strain['Replicate']
    data_strain = genes[conditions_strain['Measurment']].T[data_reference.columns]
    print(strain,data_strain.shape)
    data_strain = pd.DataFrame(scaler.transform(data_strain,log=log,scale=scale), index=data_strain.index, 
                                       columns=data_strain.columns)
    pca_transformed = pca.transform(data_strain).ravel()
    data_transformed=pd.DataFrame({'x':times,'y':pca_transformed})
    data_transformed=data_transformed.sort_values('x')
    var_sum=0
    for x,ys in data_transformed.groupby('x'):
        var_sum+=ys['y'].var()*ys.shape[0]

    #Cross validate - loo x out (all points at that x) - to find n_knots
    # For result keep MSE of highest 's' that produced that N of knots
    splitter=LeaveOneOut()
    squarred_errors=defaultdict(list)
    squarred_errors_uniqueLoo=defaultdict(list)
    x_list=data_transformed['x'].unique()
    for train_index, test_index in splitter.split(x_list):
        x_train,x_test=x_list[train_index],x_list[test_index]
        data_train=data_transformed[data_transformed['x'].isin(x_train)]
        data_test=data_transformed[data_transformed['x'].isin(x_test)]
        #data_train, data_test = data_transformed.iloc[train_index,:], data_transformed.iloc[test_index,:]
        var_sum=0
        for x,ys in data_train.groupby('x'):
            samples=ys.shape[0]
            if samples>1:
                var_sum+=ys['y'].var()*samples

        s=0.1
        first_two=None
        n_knots=np.inf
        n_knots2=0
        squarred_errors_loo=dict()
        while n_knots2<10:
            spline=UnivariateSpline(data_transformed['x'],data_transformed['y'],k=3,
                               s=var_sum*s)
            prediction=spline(data_test['x'])
            #Preediction not nan
            n_knots=len(spline.get_knots())
            if not np.isnan(prediction).all():
                if n_knots==2:
                    n_knots2+=1
                squared_error=(data_test['y']-prediction)**2
                squarred_errors[n_knots].extend(list(squared_error.values))
                squarred_errors_loo[n_knots]=list(squared_error.values)
            else:
                squarred_errors[n_knots].extend(prediction)
                squarred_errors_loo[n_knots]=list(prediction)
            s+=0.005
            
        for n_knots, squared_error in squarred_errors_loo.items():
            squarred_errors_uniqueLoo[n_knots].extend(squared_error)

    # select best n knots
    #mese=pd.DataFrame()
    #for n_knots,sqes in squarred_errors.items():
    #    n=len(sqes)
    #    # To account for extreme points in some samples use median instead of mean
    #    me=np.nanmedian(sqes)
    #    nan=np.isnan(sqes).sum()/n
    #    seme=np.nanstd(sqes)/(n**0.5)
    #    print(n_knots,len(sqes),me,me+seme,nan)
    #    mese=mese.append({'n_knots':n_knots,'mese':me+seme,'nan':nan},ignore_index=True )
    #best=mese.query('nan == 0').sort_values('mese').iloc[0,:]
    #print(best)
    #n_knots_best=int(best['n_knots'])
    
    #Select n_knots based on pvalue compared to n_knots=2
    p_vals=dict()
    for n_knots, squared_errors_list in squarred_errors_uniqueLoo.items():
        if n_knots>2 and not np.isnan(squared_errors_list).any():
            p=mannwhitneyu(squared_errors_list,squarred_errors_uniqueLoo[2])   
            p_vals[n_knots]=p
    
    if (np.array(list(p_vals.values()))<0.1).any():
        n_knots_best=min(p_vals, key=p_vals.get)
    else:
        n_knots_best=2

    var_sum=0
    for x,ys in data_transformed.groupby('x'):
        samples=ys.shape[0]
        if samples>1:
            var_sum+=ys['y'].var()*samples
    
    #Find 's' on all data that produces selected n_knots
    s=1.4
    n_knots=-np.inf
    while n_knots<n_knots_best:
        spline=UnivariateSpline(data_transformed['x'],data_transformed['y'],k=3,
                           s=var_sum*s)
        prediction=float(spline(data_transformed['x'].min()))
        #Preediction not nan
        n_knots=len(spline.get_knots())
        if n_knots<n_knots_best:
            s-=0.001


    xs = np.linspace(min(times),max(times), 1000)
    fig,ax=plt.subplots()
    ax.plot(xs, spline(xs), 'g', lw=3)
    replicates_unique=list(replicates.unique())
    cmap=plt.get_cmap('tab10').colors[:len(replicates_unique)]
    rep_colours=dict(zip(replicates_unique,cmap))
    ax.scatter(times,pca_transformed,c=[rep_colours[rep] for rep in replicates])

    ax.set_title(strain+' '+str(s)+': knots: '+str(spline.get_knots())+' target n:'+str(n_knots_best))
    display(ax)
```

```python
#GAM
plots=[]
for strain in conditions['Strain'].unique():
    #Strain data (unaveraged), transform to PCA space
    conditions_strain=conditions.query('Strain =="'+strain+'"')
    times=conditions_strain['Time']
    replicates=conditions_strain['Replicate']
    data_strain = genes[conditions_strain['Measurment']].T[data_reference.columns]
    #print(strain,data_strain.shape)
    data_strain = pd.DataFrame(scaler.transform(data_strain,log=log,scale=scale), index=data_strain.index, 
                                       columns=data_strain.columns)
    pca_transformed = pca.transform(data_strain).ravel()
    data_transformed=pd.DataFrame({'x':times,'y':pca_transformed})
    data_transformed=data_transformed.sort_values('x')
    
    #CV to select GAM parameters (regularisation, n splines)
    #CV with loo x (e.g. all points at that x used for testing)
    param_combinations=list(itertools.product(*[list(np.logspace(-5, 0, 11,base=2)),[10,20,30]]))
    param_combinations=[{'lam':lam,'n_splines':n_splines} for lam,n_splines in param_combinations]
    
    splitter=LeaveOneOut()
    squarred_errors=defaultdict(list)
    x_list=data_transformed['x'].unique()
    for train_index, test_index in splitter.split(x_list):
        x_train,x_test=x_list[train_index],x_list[test_index]
        data_train=data_transformed[data_transformed['x'].isin(x_train)]
        data_test=data_transformed[data_transformed['x'].isin(x_test)]

        for param_idx,params in enumerate(param_combinations):
            gam=pygam.LinearGAM(pygam.s(0,**params))
            gam.fit(data_train['x'].values.reshape(-1,1),data_train['y'].values.reshape(-1,1))
            prediction=gam.predict(data_test['x'].values.reshape(-1,1))
            #MSE of all points at test location
            squared_error=(data_test['y']-prediction)**2
            squarred_errors[param_idx].extend(list(squared_error.values))
            
    #Select params with smallest average MSE
    mese=pd.DataFrame()
    for param_idx,sqes in squarred_errors.items():
        me=np.nanmean(sqes)
        #print(param_combinations[param_idx],me)
        mese=mese.append({'param_idx':param_idx,'mese':me},ignore_index=True )
    best=mese.sort_values('mese').iloc[0,:]
    #print(best)
    params_best=param_combinations[int(best['param_idx'])]
    print(strain,params_best)
    
    #Make the model on whole dataset and plot it
    gam=pygam.LinearGAM(pygam.s(0,**params_best))
    gam.fit(data_transformed['x'].values.reshape(-1,1),data_transformed['y'].values.reshape(-1,1))    
    
    xs = np.linspace(min(times),max(times), 100)
    fig,ax=plt.subplots()
    ax.plot(xs, gam.predict(xs), 'r', lw=3)
    #Colour each replicate separately
    replicates_unique=list(replicates.unique())
    cmap=plt.get_cmap('tab10').colors[:len(replicates_unique)]
    rep_colours=dict(zip(replicates_unique,cmap))
    ax.scatter(times,pca_transformed,c=[rep_colours[rep] for rep in replicates])
    ax.set_title(strain)
    plots.append((fig,ax))
    display(ax)
```

```python
#Smooth with loess (+GAM)
#Loess degree
degree=3
plots=[]
for strain in conditions['Strain'].unique():
    #Strain data
    conditions_strain=conditions.query('Strain =="'+strain+'"')
    times=conditions_strain['Time']
    replicates=conditions_strain['Replicate']
    data_strain = genes[conditions_strain['Measurment']].T[data_reference.columns]
    #print(strain,data_strain.shape)
    data_strain = pd.DataFrame(scaler.transform(data_strain,log=log,scale=scale), index=data_strain.index, 
                                       columns=data_strain.columns)
    #Transform samples to PCA space
    pca_transformed = pca.transform(data_strain).ravel()
    data_transformed=pd.DataFrame({'x':times,'y':pca_transformed})
    data_transformed=data_transformed.sort_values('x')
    
    #CV loess to obtain the window (proportion of all samples to use as neighbours)
    #CV with loo x (leave all points at x as test set)
    splitter=LeaveOneOut()
    squarred_errors=defaultdict(list)
    x_list=data_transformed['x'].unique()
    for train_index, test_index in splitter.split(x_list):
        x_train,x_test=x_list[train_index],x_list[test_index]
        data_train=data_transformed[data_transformed['x'].isin(x_train)]
        data_test=data_transformed[data_transformed['x'].isin(x_test)]
        
        step=0.01
        for proportion_neigh in np.arange(0.5,0.8,step):
            loess=Loess(data_train['x'].values,data_train['y'].values,degree=degree)
            prediction=[]
            for x in data_test['x'].values:
                y=loess.estimate(x,window=int(data_train['x'].shape[0]*proportion_neigh),degree=degree)
                prediction.append(y)
            prediction=np.array(prediction)
            #MSE for all points at test x
            squared_error=(data_test['y']-prediction)**2
            squarred_errors[proportion_neigh].extend(list(squared_error.values))
            
    #Select window dataset proportion with lowest MSE
    mese=pd.DataFrame()
    for param,sqes in squarred_errors.items():
        me=np.nanmean(sqes)
        #print(param_combinations[param_idx],me)
        mese=mese.append({'param':param,'mese':me},ignore_index=True )
    best=mese.sort_values('mese').iloc[0,:]
    params_best=best['param']
    print(strain,params_best)
    
    #Fit on wholedataset and predict for plotting
    loess=Loess(data_transformed['x'].values,data_transformed['y'].values,degree=degree)
    xs = np.linspace(min(times),max(times), 100)
    prediction=[]
    for x in xs:
        y=loess.estimate(x,window=int(data_transformed['x'].shape[0]*params_best),degree=degree)
        prediction.append(y)

    fig,ax=plt.subplots(figsize=(7,4))
    ax.plot(xs, prediction, 'r', lw=6,label='loess')
    
    #Fit (and regularise) GAM over the loess predictions (stretch of points for plotting) - smooth the loess
    gam=pygam.LinearGAM(pygam.s(0))
    
    gam.gridsearch(xs.reshape(-1,1),np.array(prediction).reshape(-1,1),progress=False
                  ,lam=np.logspace(-1, 10, 11,base=2)
                   ,n_splines=[5,10,20,25]
                  )
    ax.plot(xs, gam.predict(xs), 'g', linewidth=2,label='loess+GAM')
    
    replicates_unique=list(replicates.unique())
    cmap=plt.get_cmap('tab10').colors[:len(replicates_unique)]
    rep_colours=dict(zip(replicates_unique,cmap))
    ax.scatter(times,pca_transformed,c=[rep_colours[rep] for rep in replicates],s=60)
    ax.set_title(strain)
    ax.legend()
    plots.append((fig,ax))
    display(ax)
```
