```python
import os
import sys
from importlib import reload

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.insert(0,module_path)

import matplotlib.pyplot as plt
import glob
import pandas as pd
import numpy as np
from collections import defaultdict 
from statistics import stdev, mean

from networks.library_regulons import NeighbourCalculator
from stages_DE.stages_library import PHENOTYPES
```

```python
path = '/home/karin/Documents/timeTrajectories/data/stages/images/'
dataPath = '/home/karin/Documents/timeTrajectories/data/RPKUM/combined/'
conditions_img = pd.read_csv(dataPath + 'conditions_mergedGenes_images.tsv', sep='\t', index_col=None)
```

For each image calculates 3 closest neighbours (actually 4, but one is excluded as it is itself). Comparing neighbours' phenotypes to query get average number of differences between stage annotations (annotations are 0/1 for each stage/phenotype). For each embedding calculate mean and stdev for the mean difference of each image to its neighbours. Also do this by stage; if image has 1 for a stage also annotate its mean difference to the mean differences of that stage.

```python
result=[]
dist_data={}
for embedder in ['inception-v3','SqueezeNet','VGG-16','VGG-19','Painters']:
    #print(embedder)
    embedding=pd.read_table(path+'EmbeddedImages_'+embedder+'.tab',index_col='image')

    nc=NeighbourCalculator(genes=embedding, remove_zero=False)

    neigh,sims=nc.neighbours(n_neighbours=4, inverse=False, scale='none', log=False, 
                      return_neigh_dist=True, remove_self = True)
    avg_distances=defaultdict(list)
    for image,data in neigh.iterrows():
        dist=0
        #print(image)
        phenotype=conditions_img.query('Image == "'+image+'"')[PHENOTYPES]
        #print(phenotype.values)
        for image_neigh in data:
            phenotype_neigh=conditions_img.query('Image == "'+image_neigh+'"')[PHENOTYPES]
            dist+=(phenotype_neigh.values!=phenotype.values).sum()
            #print('Dist',(phenotype_neigh.values!=phenotype.values).sum(),'sum',dist)
            #print(phenotype_neigh.values)
        avg_dist=dist/float(data.shape[0])
        avg_distances['all'].append(avg_dist)
        #print(avg_dist)
        for pheno,present in phenotype.iteritems():
            #print(pheno,present.values[0])
            if present.values[0]==1:
                avg_distances[pheno].append(avg_dist)
                #print(avg_pheno_distances)

    for phenotype,data in avg_pheno_distances.items():
        if len(data)<2:
            sd=np.nan
        else:
            sd=round(stdev(data),3)
        result.append({'Mean':round(mean(data),3),'Stdev':sd,
                   'N':len(data),'Stage':phenotype,'Embedder':embedder})
    dist_data[embedder]=avg_distances
    
```

Distribution of differences between query image and neighbour phenotypes in different embedding groups.

```python
boxplot_data=defaultdict(list)
boxplot_names=defaultdict(list)
for embedder,data in dist_data.items():
    for group,group_data in data.items():
        boxplot_names[group].append(embedder+'_'+group)
        boxplot_data[group].append(group_data)
for group in boxplot_names.keys():
    fig, ax1 = plt.subplots(figsize=(10, 5))        
    f=ax1.boxplot(boxplot_data[group])
    f=ax1.set_xticklabels(boxplot_names[group],rotation=90)
    ax1.set_title(group)
```
